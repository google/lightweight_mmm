# LightweightMMM Models.

The LightweightMMM can either be run using data aggregated at the national level (standard approach) or using data aggregated at a geo level (sub-national hierarchical approach). These models are documented below.

## National level (standard approach)

All the parameters in our Bayesian model have priors which have been set based on simulated studies that produce stable results. We also set out our three different approaches to saturation and media lagging effects: carryover (with exponent), adstock (with exponent) and hill adstock. Please see Jin, Y. et al., (2017) for more details on these models and choice of priors.

 $$kpi_{t} = \alpha + trend_{t} + seasonality_{t} + media\ channels_{t} + other\ factors_{t} $$

*Intercept*<br>
- $\alpha \sim HalfNormal(2)$

*Trend*<br>
- $trend_{t} = \mu t^{\kappa}$<br>
- $\mu \sim Normal(0,1)$<br>
- $\kappa \sim Uniform(0.5,1.5)$<br>
- Where $t$ is a linear trend input

*Seasonality (for models using* **weekly observations**)<br>
- $seasonality_{t} = \displaystyle\sum_{d=1}^{2} (\gamma_{1,d} cos(\frac{2 \pi dt}{52}) + \gamma_{2,d} sin(\frac{2 \pi dt}{52}))$<br>
- $\gamma_{1,d}, \gamma_{2,d} \sim Normal(0,1)$<br>

*Seasonality (for models using* **daily observations**)<br>
- $seasonality_{t} = \displaystyle\sum_{d=1}^{2} (\gamma_{1,d} cos(\frac{2 \pi dt}{365}) + \gamma_{2,d} sin(\frac{2 \pi dt}{365})) + \delta_{t \bmod 7}$<br>
- $\gamma_{1,d}, \gamma_{2,d} \sim Normal(0,1)$<br>
- $\delta_{i} \sim Normal(0,0.5)$<br>

*Other factors*<br>
- $other\ factors_{t} = \displaystyle\sum_{i=1}^{N} \lambda_{i}Z_{it}$<br>
- $\lambda_{i}  \sim  Normal(0,1)$<br>
- Where $Z_{i}$ are other factors and $N$ is the number of other factors.

*Media Effect*<br>
- $\beta_{m} \sim HalfNormal(v_{m})$<br>
- Where $v_{m}$ is a scalar equal to the sum of the total cost of media channel $m$.

*Media Channels (for the* **carryover** *model)*<br>
- $media\ channels_{t} = x_{t,m}^{*\rho_{m}}$<br>
- $x_{t,m}^{*} = \frac{\displaystyle\sum_{l=0}^{L} \tau_{m}^{(l-\theta_{m})^2}x_{t-l,m}}{\displaystyle\sum_{l=0}^{L}\tau_{m}^{(l-\theta_{m})^2}}$  where $L=12$<br>
- $\tau_{m} \sim Beta(1,1)$<br>
- $\theta_{m} \sim HalfNormal(2)$<br>
- $\rho_{m} \sim Beta(9,1)$
- Where $x_{t,m}$ is the media spend or impressions in week $t$ from media channel $m$<br>

*Media Channels (for the* **adstock** *model)*<br>
- $media\ channels_{t} = x_{t,m,s}^{*\rho_{m}}$<br>
- $x_{t,m,s}^{*} = \frac{x_{t,m}^{*}}{1/(1-\lambda_{m})}$<br>
- $x_{t,m}^{*} = x_{t,m} + \lambda_{m} x_{t-1,m}^{*}$ where $t=2,..,N$<br>
- $x_{1,m}^{*} = x_{1,m}$<br>
- $\lambda_{m} \sim Beta(2,1)$<br>
- $\rho_{m} \sim Beta(9,1)$
- Where $x_{t,m}$ is the media spend or impressions in week $t$ from media channel $m$<br>

*Media Channels (for the* **hill_adstock** *model)*<br>
- $media\ channels_{t} = \frac{1}{1+(x_{t,m}^{*} / K_{m})^{-S_{m}}}$<br>
- $x_{t,m}^{*} = x_{t,m} + \lambda_{m} x_{t-1,m}^{*}$  where $t=2,..,N$<br>
- $x_{1,m}^{*} = x_{1,m}$<br>
- $K_{m} \sim Gamma(1,1)$<br>
- $S_{m} \sim Gamma(1,1)$<br>
- $\lambda_{m} \sim Beta(2,1)$
- Where $x_{t,m}$ is the media spend or impressions in week $t$ from media channel $m$<br>

## Geo level (sub-national hierarchical approach)

The hierarchical model is analogous to the standard model except there is an additional dimension of region. In the geo level model seasonality is learned at the sub-national level and at the national level. For more details on this model, please see Sun, Y et al., (2017).
